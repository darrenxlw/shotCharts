```python
with open('kobe06.json') as j:
    data = json.load(j)
```

```python
kobe=pd.DataFrame(data['resultSets'][0]['rowSet'], columns=data['resultSets'][0]['headers'])
avg=pd.DataFrame(data['resultSets'][1]['rowSet'], columns=data['resultSets'][1]['headers'])
```

```python
pd.set_option('display.max_columns', None)
kobe
```

```python
#get summary of numerical data
kobe.describe()
```

```python
#check non-integer columns
intcol=[]
for c in kobe:
    if(kobe[c].dtype!="int64"):
        print("%s: " % c)
        print(kobe[c].value_counts())
        print("\n")
    else:
        intcol.append(c)
```

```python
print(kobe[intcol[1]].value_counts())
print("\n")
print(kobe[intcol[2]].value_counts())
```

No apparent erroneous data points as for non-numerical columns. Checking for outliers in numerical data in the following.

```python
#boxplot to detect outliers
kobe.boxplot(column=intcol[3:])
plt.xticks(rotation=90)

```

```python
tmp=kobe.boxplot(column="SHOT_DISTANCE", return_type="both")
```

```python
#filter dataset for outliers
kobe[kobe.SHOT_DISTANCE>=tmp.lines["whiskers"][1].get_ydata()[1]]
```

```python
kobe[kobe.LOC_Y>=kobe.boxplot(column="LOC_Y", return_type="both").lines["whiskers"][1].get_ydata()[1]]
```

Outliers is SHOT_DISTANCE and LOC_Y match up. With 0 seconds remaining in the period, these outliers can be explained as last second full-court heaves.
We now assume that the dataset is clean and without errors.
